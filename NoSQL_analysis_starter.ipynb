```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
query_count =  establishments.count_documents(query)
print(f'The number of establishments with a hygiene score of 20 is: {query_count}')
print('----------')
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(f'Number of rows:{len(results_df)}')
# Display the first 10 rows of the DataFrame
results_df.head()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
ratings_query = {'LocalAuthorityName': {'$regex': 'London'},
    'RatingValue': {'$gte': 4}
}
results = establishments.find(ratings_query)
# Use count_documents to display the number of documents in the result
query_count =  establishments.count_documents(ratings_query)
print(f'The number of establishments with with London as the Local Authority and a RatingValue greater than or equal to 4 is: {query_count}')
print('----------')
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
ratings_df= pd.DataFrame(establishments.find(ratings_query))
# Display the number of rows in the DataFrame
print(f'Number of rows:{len(ratings_df)}')
# Display the first 10 rows of the DataFrame
ratings_df.head()
```
